```python
import numpy as np
import pandas as pd
from pandas import datetime
import matplotlib.pyplot as plt
%matplotlib inline
from statsmodels.tools.eval_measures import rmse
from sklearn.preprocessing import MinMaxScaler
from keras.preprocessing.sequence import TimeseriesGenerator
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
from keras.layers import Dropout
import warnings
warnings.filterwarnings("ignore")
import pyodbc
conn_str = (
    r'DRIVER={SQL Server};'
    r'SERVER=unagisqlserver.database.windows.net;'
    r'DATABASE=unagi;'
    r'UID=Yeshan;'
    r'PWD=Un6767!#;'
)

Districts=["Ampara","Anuradhapura","Badulla","Batticoloa","Colombo","Galle","Gampaha","Hambantota","Jaffna","Kalutara","Kandy",
            "Kegalle","Kilinochchi","Kurunegala","Mannar","Matale","Matara","Monaragala","Nuwaraeliya","Polonnaruwa","Puttalam",
            "Ratnapura","Trincomalee","Vavuniya"]
for i in range(0,24):
 location=Districts[i]
 cnxn=pyodbc.connect(conn_str)
 cursor=cnxn.cursor()
 cursor.execute("SELECT Date,Temperature FROM dbo.{}".format(location))
 output=[]
 while 1:
    row=cursor.fetchone()
    if not row:
        break
    output.append(row.Date) #Appendidng the output by adding the Date column 
    output.append(row.Temperature)  #Appendidng the output by adding the Temperature column
 length=len(output)
 f=open("{}temp.csv".format(location),"w")
 f.write("Date,Temperature")
 f.write("\n")
 i=0
 while i<length:
    f.write("{},{}".format(output[i],output[i+1])) #writing to the file
    f.write("\n")
    i=i+2
 f.close()
 cnxn.close()

 def parser(x):
    return datetime.strptime(x,'%m/%Y')

 dataSet = pd.read_csv('{}temp.csv'.format(location),parse_dates=[0],date_parser=parser) #Initialising the data set
 dataSet = dataSet.set_index('Date')
 train, test = dataSet[:60], dataSet[60:]
 scaler = MinMaxScaler() #helps to normalise the ran
 scaler.fit(train) #Scaling the train data 
 train = scaler.transform(train)
 test = scaler.transform(test)
 n_input = 12 #number of time stamps
 n_features = 1 # how many features are in one time stamp
 generator = TimeseriesGenerator(train, train, length=n_input, batch_size=6)  #can have 6 samples
 model = Sequential()
 model.add(LSTM(200, activation='relu', input_shape=(n_input, n_features))) #unit which defines a vector dimension
 model.add(Dropout(0.15)) #To prevent overfitting
 model.add(Dense(1)) #Changes dimensions of the vector
 model.compile(optimizer='adam', loss='mse')
 model.fit_generator(generator,epochs=180) #runs 180 times  
 comparable_weather_predictions_list = []

 batch = train[-n_input:].reshape((1, n_input, n_features))

 for i in range(n_input):   
    comparable_weather_predictions_list.append(model.predict(batch)[0]) #We do a prediction and append it back to the list
    batch = np.append(batch[:,1:,:],[[comparable_weather_predictions_list[i]]],axis=1) #adds the previous predictions to the batch   
 predicted_weather = pd.DataFrame(scaler.inverse_transform(comparable_weather_predictions_list), #should inverse transform
                          index=dataSet[-n_input:].index, columns=['Prediction']) #creating a new column for Predictions

 
    
 testing_predictions = pd.concat([dataSet,predicted_weather], axis=1) #concatinating the predictions data frame with the data set   
 plt.figure(figsize=(20, 5))
 plt.plot(testing_predictions.index, testing_predictions['Temperature']) 
 plt.plot(testing_predictions.index, testing_predictions['Prediction'], color='r') #predicted values in red
 plt.legend(loc='best', fontsize='xx-large')
 plt.xticks(fontsize=18)
 plt.yticks(fontsize=16)
 plt.show()   
    
 pred_actual_rmse = rmse(testing_predictions.iloc[-n_input:, [0]], testing_predictions.iloc[-n_input:, [1]])
 print("rmse: ", pred_actual_rmse) #calculated the rmse


 #Starting the final predictions   
 train = dataSet #Initialising the train as the whole dataset
 scaler.fit(train)
 train = scaler.transform(train)
 n_input = 12
 n_features = 1
 generator = TimeseriesGenerator(train, train, length=n_input, batch_size=6)   
 model.fit_generator(generator,epochs=180)   
 final_weather_predictions_list = []

 batch = train[-n_input:].reshape((1, n_input, n_features))

 for i in range(n_input):   
    final_weather_predictions_list.append(model.predict(batch)[0]) 
    batch = np.append(batch[:,1:,:],[[final_weather_predictions_list[i]]],axis=1) 
 from pandas.tseries.offsets import DateOffset
 addMonths = [dataSet.index[-1] + DateOffset(months=x) for x in range(0,13) ] #Creating 12 more months
 FutureMonths = pd.DataFrame(index=addMonths[1:],columns=dataSet.columns) #Adding the new months and creating a data frame
 predicted_weather = pd.DataFrame(scaler.inverse_transform(final_weather_predictions_list), 
                          index=FutureMonths[-n_input:].index, columns=['Prediction']) #Adding the final weather predictions column
 
 final_predictions = pd.concat([dataSet,predicted_weather], axis=1) #Concat the dataset and the predicted data 
 plt.figure(figsize=(20, 5))
 plt.plot(final_predictions.index, final_predictions['Temperature'])
 plt.plot(final_predictions.index, final_predictions['Prediction'], color='r')
 plt.legend(loc='best', fontsize='xx-large')
 plt.xticks(fontsize=18)
 plt.yticks(fontsize=16)
 plt.show()   
 
 #cnxn=pyodbc.connect(conn_str)
 #cursor=cnxn.cursor()
 #arr=predicted_weather["Prediction"]
 #for i in range(1,13):
 #    print(arr[i-1])
 #val=[]
 #sql = ("INSERT INTO dbo.{} (Date,Temperature) VALUES (?,?)".format(location))
 #for i in range(1,13):
 #    if i<10:
 #        val.append((("0{}/2020").format(i),int(round(arr[i-1]))))
 #        print(val)
 #    else:
 #        val.append((("{}/2020").format(i),int(round(arr[i-1]))))
 #        print(val)
 #cursor.executemany(sql,val)
 #cnxn.commit()   
```
